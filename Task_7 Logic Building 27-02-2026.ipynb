## Problem 1: Smart Light Controller
- Problem Statement
>- Create a class to control a smart light that can be turned ON or OFF and display its status.
- 🌍 Real-Time Use
>- Smart home automation systems
- 📥 Sample Input
>- Light Name: Bedroom Light
>- Action: ON
- 📤 Expected Output
>- Bedroom Light is ON

```python
class SmartLight:
    """
    Class to control a smart light.
    Can be turned ON or OFF and display its status.
    """
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def control(self, action):
        if action.upper() in ["ON", "OFF"]:
            self.status = action.upper()
        else:
            print("Invalid Action")
            return
        print(f"{self.name} is {self.status}")
```

```python
light = SmartLight("Bedroom Light")
light.control("ON")
```

## Problem 2: Employee ID Card System
- Problem Statement
>- Create a class that stores employee name, ID, and department and displays ID card details.
- Real-Time Use
>- Corporate HR systems
- 📥 Sample Input
>- Employee Name: Rahul
>- Employee ID: EMP102
>- Department: IT
- 📤 Expected Output
>- Employee ID Card
>- Name: Rahul
>- ID: EMP102
>- Department: IT


```python
class EmployeeIDCard:
    """
    Class to store employee details and display ID card.
    """
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
```

```python
emp = EmployeeIDCard("Rahul", "EMP102", "IT")
emp.display_card()
```

## Problem 3: Mobile Contact Record
- Problem Statement
>- Create a class to store contact name and phone number and display contact information.
- Real-Time Use
>- Mobile contact applications
- 📥 Sample Input
>- Contact Name: Anita
>- Phone Number: 9876543210
- 📤 Expected Output
>- Contact Saved
>- Name: Anita
>- Phone: 9876543210

```python
class Contact:
    """
    Class to store contact details and display information.
    """
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
```

```python
contact = Contact("Anita", "9876543210")
contact.display_contact()
```

## Problem 4: Product Price Tag Generator
- Problem Statement
>- Create a class that stores product name and price and prints a formatted price tag.
-  Real-Time Use
>- Retail billing systems
- 📥 Sample Input
>- Product Name: Headphones
>- Price: 2499
- 📤 Expected Output
>- Product: Headphones
>- Price: ₹2499

```python
class Product:
    """
    Class to store product details and print price tag.
    """
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
```

```python
item = Product("Headphones", 2499)
item.display_price_tag()
```

## Problem 6: Delivery Address Manager
- Problem Statement
>- Create a class that stores customer name and delivery address and prints delivery details.
- Real-Time Use
>- Courier and logistics platforms
- 📥 Sample Input
>- Customer Name: Suman
>- Address: Hyderabad
- 📤 Expected Output
>- Delivery Details
>- Customer: Suman
>- Address: Hyderabad

```python
class DeliveryAddress:
    """
    Class to store customer delivery details.
    """
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display_delivery(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")
```

```python
delivery = DeliveryAddress("Suman", "Hyderabad")
delivery.display_delivery()
```
