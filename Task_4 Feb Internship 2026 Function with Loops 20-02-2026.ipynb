## Problem 1: Mobile Recharge Validation System
- Problem Statement
>- Create a function to validate a mobile recharge request.
- Rules:
>- Recharge amount must be ≥ ₹50
>- Valid plans: 199, 299, 399, 599
>- User should be allowed to retry if invalid

- Real-Time Use
>- Telecom recharge apps
>- Payment gateways

- Hint
>- Use function for validation
>- Use while loop for retry

```python
def validate_recharge():
    """
    Function to validate mobile recharge requests.
    Rules:
    - Minimum recharge amount: ₹50
    - Valid plans: 199, 299, 399, 599
    - Retry allowed until valid input is given
    """
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: "))
            if amount >= 50 and (amount in valid_plans or amount >= 50):
                print("Recharge Successful..!")
                break
            else:
                print("Invalid recharge amount. Please try again.")
        except ValueError:
            print("Invalid input. Enter a numeric value.")
```

## Problem 2: Inventory Reorder Alert System
- Problem Statement
>- Given a dictionary of products and stock quantities:
>- Print Reorder Alert if stock < 15
>- Print Stock OK otherwise
- Real-Time Use
>- Warehouse systems
>- Retail management
- Hint
>- Use function
>- Use for loop on dictionary

```python
def check_inventory(stock_dict):
    """
    Function to check inventory levels.
    Prints 'Reorder Alert' if stock < 15, else 'Stock OK'.
    """
    for product, qty in stock_dict.items():
        if qty < 15:
            print(f"{product}: Reorder Alert  (Stock = {qty})")
        else:
            print(f"{product}: Stock OK (Stock = {qty})")


# Example usage
inventory = {"Apples": 10, "Bananas": 25, "Oranges": 12}
check_inventory(inventory)
```

## Problem 3: Student Result Processing System
- Problem Statement
>- Create a function that:
>- Accepts a list of student marks
>- Calculates average marks
>- Prints Pass if average ≥ 50 else Fail
- Real-Time Use
>- Online exam systems
>- Evaluation portals
- Hint
>- Use loop to calculate total
>- Return result from function

```python
def process_result(marks):
    """
    Function to calculate average marks and print Pass/Fail.
    Pass if average ≥ 50, else Fail.
    """
    total = 0
    for mark in marks:
        total += mark
    
    avg = total / len(marks)
    print(f"Average Marks = {avg:.2f}")
    
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"


# Example usage
marks_list = [60, 45, 70, 55]
print(process_result(marks_list))
```

## Problem 4: Cab Fare Estimator with Retry Option
- Problem Statement
>- Create a cab fare calculator:
>- Base fare = ₹50
>- ₹12 per km
>- Peak hour → 25% extra
>- Allow retry using while loop
- Real-Time Use
>- Ride-hailing apps
>- Transport systems
- Hint
>- Function for fare calculation
>- While loop for retry

```python
def calculate_fare(distance, peak_hour=False):
    """
    Function to calculate cab fare.
    Base fare = ₹50
    ₹12 per km
    Peak hour → 25% extra
    """
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak_hour:
        fare *= 1.25  # Adding 25% extra

    return fare


def cab_fare_system():
    while True:
        try:
            km = float(input("Enter distance travelled (in km): "))
            peak = input("Is it peak hour? (yes/no): ").lower() == "yes"
            
            total_fare = calculate_fare(km, peak)
            print(f"Total Fare = ₹{total_fare:.2f}")
            break
        except ValueError:
            print("Invalid input. Please try again.")
```

```python
cab_fare_system()
```

## Problem 5: Employee Attendance Eligibility Checker
- Problem Statement
>- Create a function that:
>- Accepts attendance list ("P", "A")
>- Calculates attendance percentage
>- Returns Eligible if ≥ 75% else Not Eligible
- Real-Time Use
>- HR payroll systems
>- Internship tracking
- Hint
>- Use loop to count present days
>- Use return statement

```python
def check_attendance(attendance_list):
    """
    Function to check employee attendance eligibility.
    Eligible if attendance ≥ 75%, else Not Eligible.
    """
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    print(f"Attendance Percentage = {percentage:.2f}%")
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
print(check_attendance(attendance))
```

## Problem 6: Password Strength Checker
- Problem Statement
>- Create a function to check password strength.
>- Rules:
>- Minimum length = 8
>- Must contain at least one digit
>- Must contain at least one special character (@#$)
- Real-Time Use
>- Login systems
>- Security validation
- Hint
>- Use loop over string
>- Use conditions inside function

```python
def check_password_strength(password):
    """
    Function to check password strength.
    Rules:
    - Minimum length = 8
    - Must contain at least one digit
    - Must contain at least one special character (@,#,$ etc.)
    """
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak (Too short)"
    
    if not any(char.isdigit() for char in password):
        return "Weak (No digit found)"
    
    if not any(char in special_chars for char in password):
        return "Weak (No special character found)"
    
    return "Strong "
```
